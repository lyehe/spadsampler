```python
data = np.random.randint(0, 256, (100, 100, 3), dtype=np.uint8)
p = np.ones_like(data) * 0.5
p = p.mean(axis=(0, 1), keepdims=True)
binomial_sample = np.random.binomial(data, p)

print(data.shape, p.shape, binomial_sample.shape)
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 2)
ax[0].imshow(data)
ax[1].imshow(binomial_sample)
```
